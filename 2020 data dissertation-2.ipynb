```python
#First import all the libraries we will be using for analysis:
import numpy as np
import pandas as pd
import scipy.stats as stats
from pathlib import Path
from pandas.api.types import CategoricalDtype
#and for plotting graphs:
import matplotlib.pyplot as plt
import seaborn as sns
#and set up the figure parameters
plt.rcParams['figure.figsize'] = [12, 7]
sns.set()
sns.set_style("whitegrid")
sns.set_palette("bright")
```

```python
# Import the database into a dataframe
data_file =('PERFECT-Diss-Data.xlsx')
df_Gender_Climate = pd.read_excel(data_file, index_col=0)
```

```python
df_Gender_Climate['GGI 2020'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['EPI 2020'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['GII 2019'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['CCH EPI 2020'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['GHP EPI 2020'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['GIB EPI 2020'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['GGIE 2020'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['GGIEA 2020'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['GGIH 2020'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['GGIP 2020'].replace(to_replace=[999], value=np.nan, inplace=True)
```

```python
df_Gender_nonans = df_Gender_Climate.dropna()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 4))
rp_GGI = stats.pearsonr(df_Gender_nonans['GGI 2020'], df_Gender_nonans['EPI 2020'])
rp_GII = stats.pearsonr(df_Gender_nonans['GII 2019'], df_Gender_nonans['EPI 2020'])

sns.regplot(ax=ax[0], data=df_Gender_nonans, x='EPI 2020', y='GGI 2020', color = 'blue')
ax[0].set_ylabel('GGGI')
ax[0].set_xlabel('EPI')
ax[0].set_title('(a) Correlations between GGGI and EPI')
ax[0].text(0.6, 0.2, "Pearson R = {:.2f}\np = {:.2f}".format(rp_GGI[0],rp_GGI[1]), transform=ax[0].transAxes)

sns.regplot(ax=ax[1], data=df_Gender_nonans, x='EPI 2020', y='GII 2019', color = 'green') 
ax[1].set_ylabel('GII')
ax[1].set_xlabel('EPI')
ax[1].set_title('(b) Correlations between GII and EPI')
ax[1].text(0.6, 0.8, "Pearson R = {:.2f}\np = {:.2f}".format(rp_GII[0],rp_GII[1]), transform=ax[1].transAxes)

plt.savefig('H1_graph')
```
