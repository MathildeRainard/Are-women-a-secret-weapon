```python
#First import all the libraries we will be using for analysis:
import numpy as np
import pandas as pd
import scipy.stats as stats
from pathlib import Path
from pandas.api.types import CategoricalDtype
#and for plotting graphs:
import matplotlib.pyplot as plt
import seaborn as sns
#and set up the figure parameters
plt.rcParams['figure.figsize'] = [12, 7]
sns.set()
sns.set_style("whitegrid")
sns.set_palette("bright")
```

```python
# Import the database into a dataframe
data_file =('Diss_data_2010_GII_GGI_EPI.xlsx')
df_Gender_Climate = pd.read_excel(data_file, index_col=0)
```

```python
df_Gender_Climate['GGI 2010'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['EPI 2010'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['GII 2010'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['CCH EPI 2010'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['GHP EPI 2010'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate['GIB EPI 2010'].replace(to_replace=[999], value=np.nan, inplace=True)
```

```python
df_Gender_nonans = df_Gender_Climate.dropna()
```

```python
stats.pearsonr(df_Gender_nonans['GGI 2010'], df_Gender_nonans['EPI 2010'])
```

```python
stats.pearsonr(df_Gender_nonans['GII 2010'], df_Gender_nonans['EPI 2010'])
```

```python
stats.pearsonr(df_Gender_nonans['GGI 2010'], df_Gender_nonans['GIB EPI 2010'])
```

```python
stats.pearsonr(df_Gender_nonans['GII 2010'], df_Gender_nonans['GIB EPI 2010'])
```

```python
stats.spearmanr(df_Gender_nonans['GII 2010'], df_Gender_nonans['GIB EPI 2010'])
```

```python
stats.pearsonr(df_Gender_nonans['GII 2010'], df_Gender_nonans['CCH EPI 2010'])
```

```python
stats.pearsonr(df_Gender_nonans['GGI 2010'], df_Gender_nonans['CCH EPI 2010'])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 4))
  
fig.suptitle('EPI correlation with GGI and GII')

sns.regplot(ax=ax[0], data=df_Gender_nonans, x='EPI 2010', y='GGI 2010', color = 'blue')
ax[0].set_ylabel('Gender Gap Index')
ax[0].set_xlabel('EPI')


sns.regplot(ax=ax[1], data=df_Gender_nonans, x='EPI 2010', y='GII 2010', color = 'green') 
ax[1].set_ylabel('Gender Inequality Index')
ax[1].set_xlabel('EPI')
```

```python
data_file =('DISS-DATA-2010-WB-Regions.xlsx')
df_Gender_Climate_Regions = pd.read_excel(data_file, index_col=0)
```

```python
df_Gender_Climate_Regions['GII 2010 LW'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['GII 2010 LM'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['GII 2010 UM'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['GII 2010 HG '].replace(to_replace=[999], value=np.nan, inplace=True)

df_Gender_Climate_Regions['GGI 2010 LW'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['GGI 2010 LM'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['GGI 2010 UM'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['GGI 2010 HG'].replace(to_replace=[999], value=np.nan, inplace=True)

df_Gender_Climate_Regions['EPI 2010 LW'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['EPI 2010 LM'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['EPI 2010 UM'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['EPI 2010 HG'].replace(to_replace=[999], value=np.nan, inplace=True)

df_Gender_Climate_Regions['CCH EPI 2010 LW'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['CCH EPI 2010 LM'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['CCH EPI 2010 UM'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['CCH EPI 2010 HG'].replace(to_replace=[999], value=np.nan, inplace=True)

df_Gender_Climate_Regions['GHP EPI 2010 LW'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['GHP EPI 2010 LM'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['GHP EPI 2010 UM'].replace(to_replace=[999], value=np.nan, inplace=True)
df_Gender_Climate_Regions['GHP EPI 2010 HG'].replace(to_replace=[999], value=np.nan, inplace=True)
```

```python
fig, ax = plt.subplots(1, 4, figsize=(20, 5))
  
fig.suptitle('GGI and EPI per income region 2010')

sns.scatterplot(ax=ax[0], data=df_Gender_Climate_Regions, x='EPI 2010 LW', y='GGI 2010 LW', color = 'blue')
ax[0].set_title("Low Income", fontsize=12)
ax[0].set_ylabel('Gender Gap Index')
ax[0].set_xlabel('EPI')

sns.scatterplot(ax=ax[1], data=df_Gender_Climate_Regions, x='EPI 2010 LM', y='GGI 2010 LM', color = 'blue')
ax[1].set_title("Lower-middle Income", fontsize=12)
ax[1].set_ylabel('Gender Gap Index')
ax[1].set_xlabel('EPI')

sns.scatterplot(ax=ax[2], data=df_Gender_Climate_Regions, x='EPI 2010 UM', y='GGI 2010 UM', color = 'blue')
ax[2].set_title("Upper-middle Income", fontsize=12)
ax[2].set_ylabel('Gender Gap Index')
ax[2].set_xlabel('EPI')

sns.regplot(ax=ax[3], data=df_Gender_Climate_Regions, x='EPI 2010 HG', y='GGI 2010 HG', color = 'blue')
ax[3].set_title("High Income", fontsize=12)
ax[3].set_ylabel('Gender Gap Index')
ax[3].set_xlabel('EPI')
```
